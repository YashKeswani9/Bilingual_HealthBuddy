```python
import csv
import json

# Input CSV file path
csv_file_path = "/kaggle/input/medquad/medDataset_processed.csv" 
json_file_path = "medquad_preprocessed.json"

data = []

# Read CSV and transform it into JSON
with open(csv_file_path, mode="r", encoding="utf-8") as f:
    reader = csv.DictReader(f)  # Assumes first row contains headers
    for row in reader:
        entry = {
            "question": row["Question"],  # Extract question
            "answer": row["Answer"],      # Extract answer
        }
        data.append(entry)

# Save to JSON file
with open(json_file_path, mode="w", encoding="utf-8") as f:
    json.dump(data, f, indent=4)

print(f"✅ JSON data saved to {json_file_path}")
```

```python
import pandas as pd

# Load the CSV file (replace 'your_file.csv' with the actual file path)
df = pd.read_csv("/kaggle/input/medquad/medDataset_processed.csv")

# Ensure the 'question' column exists
if "Question" not in df.columns:
    raise ValueError("Column 'question' not found in the CSV file. Please check the file format.")

# Extract all unique questions
questions = df["Question"].dropna().unique()

# Prepare the YAML formatted string
yaml_content = "- intent: ask_health_question\n  examples: |\n"
for q in questions:
    yaml_content += f"    - {q}\n"

# Save to `nlu.yml`
with open('/kaggle/working/nlu.yml', 'w') as file:
    file.write("\n" + yaml_content)

print(f"✅ Successfully added {len(questions)} questions to nlu.yml!")
```
